5). Carry out min-max scaling of the exploratory variables. HINT: use preprocessing.MinMaxScaler().

```python
X = data.iloc[:, 0:53]
y = data.iloc[:, 54]

X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
X_train  = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

6). Choose an algorithm and carry out the predictive analysis.

- Optimize the hyperparameter(s)

```python
XGBC=GradientBoostingClassifier(learning_rate=0.1, max_depth=2, n_estimators=300,random_state=0)
XGBC.fit(X_train, y_train)
Y_pred = XGBC.predict(X_test)
print( "Gradient Boosting accuracy : " + str(np.round(metrics.accuracy_score(y_test,Y_pred),3)))
```
